```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import scipy.stats as stats
import scipy.cluster.hierarchy as sch
from operator import *
pd.core.config.option_context('mode.use_inf_as_null',True)
import seaborn
rcdefsns = plt.rcParams.copy()
import brewer2mpl
import os 
import sys
```

```python
data=pd.read_table('data_1',header=0)
data['Read_length']=data['Read'].apply(len)
data['Score']=data['Score'].apply(lambda x:int(x.split(':')[-1]))
data=data[data['Read_length']>1500]
data.sort(columns=['Read_id','Score'], inplace=True, ascending=[False,False])
data=data.drop_duplicates(subset=['Read_id'])
```

```python
len(data)
```

```python
report=data[['Read_id','tName_x','Score','Read_length','matches','misMatches','qBaseInsert','tBaseInsert','percentIdentity']]
```

```python
report.rename(columns={'tName_x': 'Isoform', \
                       'Score': 'LAST_Score',\
                       'matches': 'Matches', \
                       'misMatches':'Mismatches',\
                       'qBaseInsert':'Insertions',\
                       'tBaseInsert':'Deletions',\
                       'percentIdentity':'Percent_Identity'\
                       },inplace=True)
```

```python
report.to_csv('Data_Table',sep='\t')
```

```python
report['direction']=report['Read_id'].apply(lambda x : x.split('_')[-1])
```

```python
matplotlib.rcParams['figure.figsize'] = (10.0, 5.0)
figure=plt.figure()
ax=plt.subplot2grid((1,2),(0,0),rowspan=1,colspan=1)
seaborn.violinplot(report.Percent_Identity,report.direction,ax=ax)
ax.tick_params(axis='both', which='major', labelsize=12)
ax.set_title('Percent Identity',fontsize=18)
ax.set_xlabel('')
ax.set_ylabel('Percent Identity')
ax.tick_params(axis='both', which='major', labelsize=14)

ax=plt.subplot2grid((1,2),(0,1),rowspan=1,colspan=1)
seaborn.violinplot(report.LAST_Score,report.direction,ax=ax)
ax.tick_params(axis='both', which='major', labelsize=12)
ax.set_title('LAST Score',fontsize=18)
ax.set_xlabel('')
ax.set_ylabel('LAST Score')
ax.tick_params(axis='both', which='major', labelsize=14)
```

```python
figure.savefig('Full_Length_Score_PID.pdf',format='pdf',dpi=600)
```
