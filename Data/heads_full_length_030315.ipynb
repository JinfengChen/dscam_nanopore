```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import scipy.stats as stats
import scipy.cluster.hierarchy as sch
from operator import *
pd.core.config.option_context('mode.use_inf_as_null',True)
import seaborn
rcdefsns = plt.rcParams.copy()
import brewer2mpl
```

```python
align=pd.read_table('SAM/dscam_heads_Isoforms.sam',index_col=False,names=['Read_id','Flag','Align_Exon','Start','B','CIGAR','A','P','P2','Read','A2','Mismatches','Score'], skiprows=1)
alignpsl=pd.read_table('PSL/dscam_heads_Isoforms.psl',index_col=False,names=['matches','misMatches','repMatches','nCount','qNumInsert','qBaseInsert','tNumInsert','tBaseInsert','strand','qName','qSize','qStart','qEnd','tName','tSize','tStart','tEnd','blockCount','blockSizes','qStarts','tStarts'])
```

```python
alignpsl.head()
```

```python
align.drop(['Flag','B','A','P','P2','A2'],axis=1,inplace=True)
```

```python
print len(align)
```

```python
align['Align_Exon']=align['Align_Exon'].apply(lambda x:x.replace('Dscam_',''))
align['Align_Exon']=align['Align_Exon'].apply(lambda x:'.'.join(x.split('_')[::-1]))
#align['Score']=align['Score'].apply(lambda x:int(x.split(':')[-1]))
align['Length']=align['Read'].apply(len)
align.head()
```

```python
count_appkey = align.groupby('Read_id')['Read_id'].transform('count')
align=align[count_appkey==1]
count_appkey = alignpsl.groupby('qName')['qName'].transform('count')
alignpsl=alignpsl[count_appkey==1]
```

```python
align.set_index('Read_id',inplace=True)
alignpsl.set_index('qName',inplace=True)
alignpsl=alignpsl[alignpsl.columns[:3]]
```

```python
inta=align[align['Length']>1500]['Align_Exon'].value_counts()
inta[inta<20].sum()
```

```python
align['Length'].plot(kind='hist',bins=np.arange(0,2500,100),normed=True)
```

```python
align['Start'].plot(kind='hist',bins=np.arange(0,2500,100))
```

```python
alldata=align.join(alignpsl).dropna()
```

```python
alldata.head()
```

```python
alldata.reset_index(inplace=True)
alldata['direction']=alldata['Read_id'].apply(lambda x: x.split(' ')[0].split('_')[-1])
alldata['id']=alldata['Read_id'].apply(lambda x: x.split(' ')[0])
```

```python
alldata.set_index('Read_id',inplace=True)
```

```python
alldata.head()
```

```python
alldata['PercentIdentity']=(alldata['matches']*100.)/(alldata['matches']+alldata['misMatches'])
filterdata=alldata[alldata['Length']>1500]
```

```python
matplotlib.rcParams['figure.figsize'] = (5.0, 5.0)
figure=plt.figure()
ax=figure.add_subplot(111)
filterdata['PercentIdentity'].plot(kind='hist',bins=np.arange(0,100,5),ax=ax)
ax.set_title('Percent Identity', fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=14)
```

```python
filterdata[filterdata['PercentIdentity']==filterdata['PercentIdentity'].max()]
```

```python
g = seaborn.FacetGrid(filterdata, col="direction")
g.map(plt.hist, 'PercentIdentity', bins=np.arange(0,100,5))
g.set_xlabels('Percent Identity')
```

```python
matplotlib.rcParams['figure.figsize'] = (10.0, 10.0)
fig=plt.figure()
ax=plt.subplot2grid((2,2),(0,0),rowspan=1,colspan=1)
filterdata.groupby('direction')['PercentIdentity'].plot(kind='kde', ax=ax)
ax.legend()
for i in ax.lines:
    i.set_linewidth(3)
ax.set_xlim([0,100])
ax.set_xlabel('Percent Identity',fontsize=16)
ax=plt.subplot2grid((2,2),(0,1),rowspan=1,colspan=1)
filterdata.groupby('direction')['Length'].plot(kind='kde',ax=ax)
for i in ax.lines:
    i.set_linewidth(3)

ax.set_xlabel('Alignment Length',fontsize=16)
ax.legend()
```
